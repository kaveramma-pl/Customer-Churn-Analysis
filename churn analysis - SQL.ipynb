```python
-- View first 5 rows
SELECT *
FROM "Telco_customer_churn"
LIMIT 5;

-- Check duplicates
SELECT CustomerID, COUNT(*)
FROM "Telco_customer_churn"
GROUP BY CustomerID
HAVING COUNT(*) > 1;

-- Total rows
SELECT COUNT(*) AS total_rows
FROM "Telco_customer_churn";

-- Statistics
SELECT
    AVG(monthly_charges) AS avg_monthly_charges,
    MIN(monthly_charges) AS min_monthly_charges,
    MAX(monthly_charges) AS max_monthly_charges,
    AVG(total_charges) AS avg_total_charges
FROM "Telco_customer_churn";

-- Create cleaned table
DROP TABLE IF EXISTS Cleaned_telco_data;

CREATE TABLE Cleaned_telco_data (
    customerID TEXT,
    gender TEXT,
    contract TEXT,
    payment_method TEXT,
    monthly_charges REAL,
    total_charges REAL,
    tenure_months INTEGER,
    churn_label TEXT,
    churn_value INTEGER
);

-- Insert cleaned data
INSERT INTO Cleaned_telco_data
SELECT
    customerID,
    gender,
    contract,
    payment_method,
    monthly_charges,
    total_charges,
    tenure_months,
    churn_label,
    churn_value
FROM "Telco_customer_churn";

SELECT *
FROM Cleaned_telco_data;
```
